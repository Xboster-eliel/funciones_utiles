```python
import pandas as pd
import chardet

# Detectar la codificación del archivo CSV
with open('/content/datos_extraidos.csv', 'rb') as f:
    result = chardet.detect(f.read(100000))  # Analiza los primeros 100 KB
    encoding_detected = result['encoding']
    print("Codificación detectada:", encoding_detected)

# Cargar el CSV con la codificación detectada
df = pd.read_csv('/content/datos_extraidos.csv', encoding=encoding_detected, delimiter=';')
```

```python
# Seleccionar las columnas deseadas
df3 = df[['uu_id', 'longitude', 'latitude', 'fechano', 'fechafec', 'causa8', 'descausa','evento', 'name0', 'name1', 'lugar', 'name2','muertos', 'heridos', 'desaparece', 'afectados', 'di_comments', 'eef007', 'fuentes', 'di_comments']]

# Aplicar la limpieza a df2
df3_clean = df3.apply(lambda col: col.dropna().reset_index(drop=True))

# Guardar el archivo limpio
df3_clean.to_csv('/content/prueba2.csv', index=False, encoding=encoding_detected, sep=';')

# Mostrar las primeras filas del dataframe limpio
print(df3_clean.head())
```
